```python
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#from google.colab import drive
#drive.mount('/content/drive')
#df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Homework3 - TSLA.csv')
#df

df = pd.read_csv('Homework3 - TSLA.csv')

missing_values = df.isnull().sum()
print(missing_values)

df['Date'] = pd.to_datetime(df['Date'])
df['Date']

df.set_index('Date', inplace = True)
df.head()

plt.figure(figsize=(13,6))
plt.plot(df['Close'], label='Tesla Closing Price')
plt.title('Tesla Stock Clossing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.savefig('Tesla.png')
plt.show()

def preprocess_data(df, feature='Close', split_ratio=0.7):
  scaler = MinMaxScaler()
  scaled_data = scaler.fit_transform(df[[feature]])
  split_index = int(len(df) * split_ratio)
  train_data = scaled_data[:split_index]
  test_data = scaled_data[split_index:]
  return train_data, test_data, scaler

train_data, test_data, scaler = preprocess_data(df)

def create_sequences(df, sequence_lengh=60):
  sequences = []
  labels = []
  for i in range(len(df) - sequence_lengh):
    sequences.append(df[i:i+sequence_lengh])
    labels.append(df[i+sequence_lengh])
  return np.array(sequences), np.array(labels)

sequence_lengh = 60
X_train, y_train = create_sequences(train_data, sequence_lengh)
X_test, y_test = create_sequences(test_data, sequence_lengh)

print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)

def build_lstm_model():
  model = Sequential([
      LSTM(50,return_sequences=True,input_shape=(sequence_lengh,1)),
      Dropout(0.2),
      LSTM(50,return_sequences=False),
      Dropout(0.2),
      Dense(25),
      #Droupout(0.2),
      Dense(1)
  ])
  return model

lstm_model = build_lstm_model()
lstm_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

es = EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True)
lstm_history = lstm_model.fit(
    X_train, y_train,
    batch_size=32,
    epochs=50,
    validation_data=(X_test, y_test),
    callbacks=[es]
)

predicted = lstm_model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted)

plt.figure(figsize=(13,6))
plt.plot(df.index[-len(y_test):],scaler.inverse_transform(y_test.reshape(-1,1)),
         label='True Prices')
plt.plot(df.index[-len(y_test):],predicted_prices,
         label='Predicted Prices')
plt.title('LSTM Model Predictions or Tesla Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.savefig('LSTM.png')
plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted))
print(f"RMSE: {rmse}")

####################### Gold

! pip install yfinance

import yfinance as yf
import pandas as pd

# Define the ticker symbol for Gold Futures
gold_ticker = "GC=F"

# Define the date range (adjust based on your Tesla dataset)
start_date = "2010-01-01"  # Replace with your Tesla dataset's start date
end_date = "2022-12-12"    # Replace with your Tesla dataset's end date

# Fetch historical data for Gold Futures
gold_data = yf.download(gold_ticker, start=start_date, end=end_date)

# Save the data to a CSV file
output_file = "gold_prices.csv"
gold_data.to_csv(output_file)

print(f"Gold Futures data saved to {output_file}")

df2 = pd.read_csv('gold_prices.csv')
df2

missing_values = df2.isnull().sum()
print(missing_values)

df_cleaned = df2.dropna()

# Save the cleaned DataFrame back to a CSV file
df_cleaned.to_csv('cleaned_gold_prices.csv', index=False)

# Display the cleaned DataFrame
print(df_cleaned.head())

df3 = pd.read_csv('cleaned_gold_prices.csv')  # Read the file, treating the first row as the header

# Drop the second line (index 1) by using the `drop` method
df_cleaned2 = df3.drop(index=0)

# Save the cleaned DataFrame back to a CSV file
df_cleaned2.to_csv('cleaned_2_gold_prices.csv', index=False)

# Display the cleaned DataFrame
print(df_cleaned2.head())

df_cleaned2.rename(columns={'Price': 'Date'}, inplace=True)

# Save the cleaned DataFrame back to a CSV file
df_cleaned2.to_csv('ready_cleaned_gold_prices.csv', index=False)

# Display the DataFrame to verify the change
print(df_cleaned2.head())


from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#from google.colab import drive
#drive.mount('/content/drive')
df4 = pd.read_csv('ready_cleaned_gold_prices.csv')
df4

missing_values = df.isnull().sum()
print(missing_values)

df4['Date'] = pd.to_datetime(df4['Date'])
df4['Date']

df4.set_index('Date', inplace = True)
df4.head()

plt.figure(figsize=(13,6))
plt.plot(df4['Close'], label='Gold Closing Price')
plt.title('Gold Stock Clossing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.savefig('Gold.png')
plt.show()

def preprocess_data(df, feature='Close', split_ratio=0.7):
  scaler = MinMaxScaler()
  scaled_data = scaler.fit_transform(df[[feature]])
  split_index = int(len(df4) * split_ratio)
  train_data = scaled_data[:split_index]
  test_data = scaled_data[split_index:]
  return train_data, test_data, scaler

train_data, test_data, scaler = preprocess_data(df4)

def create_sequences(df4, sequence_lengh=60):
  sequences = []
  labels = []
  for i in range(len(df4) - sequence_lengh):
    sequences.append(df4[i:i+sequence_lengh])
    labels.append(df4[i+sequence_lengh])
  return np.array(sequences), np.array(labels)

sequence_lengh = 60
X_train, y_train = create_sequences(train_data, sequence_lengh)
X_test, y_test = create_sequences(test_data, sequence_lengh)

print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)

def build_lstm_model():
  model = Sequential([
      LSTM(50,return_sequences=True,input_shape=(sequence_lengh,1)),
      Dropout(0.2),
      LSTM(50,return_sequences=False),
      Dropout(0.2),
      Dense(25),
      #Droupout(0.2),
      Dense(1)
  ])
  return model

lstm_model = build_lstm_model()
lstm_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

es = EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True)
lstm_history = lstm_model.fit(
    X_train, y_train,
    batch_size=32,
    epochs=50,
    validation_data=(X_test, y_test),
    callbacks=[es]
)

predicted = lstm_model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted)

plt.figure(figsize=(13,6))
plt.plot(df4.index[-len(y_test):],scaler.inverse_transform(y_test.reshape(-1,1)),
         label='True Prices')
plt.plot(df.index[-len(y_test):],predicted_prices,
         label='Predicted Prices')
plt.title('Gold Model Predictions or Gold Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.savefig('Gold_2.png')
plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted))
print(f"RMSE for Gold: {rmse}")


```

```python
###### TSLA Transformers model
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam

df = pd.read_csv('Homework3 - TSLA.csv')

#file_path = '/path/to/your/new/file.csv'  # Replace with your file path
#df = pd.read_csv(file_path)

def build_transformer_model(input_shape):
    inputs = layers.Input(shape=input_shape)

    # Example: Basic Transformer-like block using MultiHeadAttention
    transformer_block = layers.MultiHeadAttention(num_heads=4, key_dim=64)(inputs, inputs)
    ff_layer = layers.Dense(128, activation='relu')(transformer_block)
    dropout_layer = layers.Dropout(0.2)(ff_layer)
    global_avg_pooling = layers.GlobalAveragePooling1D()(dropout_layer)

    outputs = layers.Dense(1)(global_avg_pooling)
    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return model

# Model configuration
sequence_length = 60
transformer_model = build_transformer_model((sequence_length, 1))

# Compile the model
transformer_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Display the model architecture
transformer_model.summary()


es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
transformer_history = transformer_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=32,
    callbacks=[es]
)


Predicted = transformer_model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted)

plt.figure(figsize=(13,6))
plt.plot(df.index[-len(y_test):],scaler.inverse_transform(y_test.reshape(-1,1)),
         label='True Prices')
plt.plot(df.index[-len(y_test):],predicted_prices,
         label='Predicted Prices')
plt.title('Transformer Model Predictions')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.savefig('Transformer_Model_Predictions.png')
plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted))
print(f"RMSE for tesla Transformers: {rmse}")

##################### Gold Transformers model
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam

df4 = pd.read_csv('ready_cleaned_gold_prices.csv')

def build_transformer_model(input_shape):
    inputs = layers.Input(shape=input_shape)

    # Example: Basic Transformer-like block using MultiHeadAttention
    transformer_block = layers.MultiHeadAttention(num_heads=4, key_dim=64)(inputs, inputs)
    ff_layer = layers.Dense(128, activation='relu')(transformer_block)
    dropout_layer = layers.Dropout(0.2)(ff_layer)
    global_avg_pooling = layers.GlobalAveragePooling1D()(dropout_layer)

    outputs = layers.Dense(1)(global_avg_pooling)
    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return model

# Model configuration
sequence_length = 60
transformer_model = build_transformer_model((sequence_length, 1))

# Compile the model
transformer_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Display the model architecture
transformer_model.summary()


es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
transformer_history = transformer_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=32,
    callbacks=[es]
)


Predicted = transformer_model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted)

plt.figure(figsize=(13,6))
plt.plot(df4.index[-len(y_test):],scaler.inverse_transform(y_test.reshape(-1,1)),
         label='True Prices')
plt.plot(df4.index[-len(y_test):],predicted_prices,
         label='Predicted Prices')
plt.title('Transformer Gold Model Predictions')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.savefig('Transformer_Gold_Model_Predictions.png')
plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted))
print(f"RMSE for gold prices transformers : {rmse}")

```
